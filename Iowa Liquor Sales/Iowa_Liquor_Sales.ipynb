# Data Cleaning

## Check for inconsistency in the City column

```python
df['City'] = df['City'].str.upper().str.strip()       # Convert city names to upper cases and strip white spaces
city = pd.DataFrame(df['City'].value_counts(ascending=True))         # Count city names
city['Name'] = city.index      # Add a new column called Name
city
```

```python
print('There are some cities with significantly low occurance, \nthis may due to typing errors, we need to double check this.')
```

```python
import nltk
q1 = city['City'].quantile(0.25)        # Get the first quartile value in the city count
for city_name in city[city['City'] < q1]['Name']:
    for compare in city['Name']:
        distance = nltk.edit_distance(compare, city_name)     # Calculate the edit distance between the city names
        if distance > 0 and distance < 3:
            print(city[city.Name == compare]['City'])
            print(city[city.Name == city_name]['City'])
            print('---------------------------------------')
```

```python
# Based on the result, we define the city name map
city_name_map = {}
for city_name in city['Name']:
    city_name_map[city_name] = city_name

city_name_map['ALVORD'] = 'WALFORD'
city_name_map['CLEARLAKE'] = 'CLEAR LAKE'
city_name_map['GRAND MOUNDS'] = 'GRAND MOUND'
city_name_map['OTUMWA'] = 'OTTUMWA'

df['City'] = df['City'].map(city_name_map)
```

## Check for inconsistency in the Vendor Name column

```python
df['Vendor Name'] = df['Vendor Name'].str.upper().str.strip()       # Convert vendor names to upper cases and strip white spaces
vendor_name = pd.DataFrame(df['Vendor Name'].value_counts(ascending=True))         # Count vendor names
vendor_name['Name'] = vendor_name.index      # Add a new column called Name
vendor_name
```

```python
q2 = vendor_name['Vendor Name'].quantile(0.5)        # Get the second quartile value in the vendor name count
for vendor in vendor_name[vendor_name['Vendor Name'] < q2]['Name']:
    for compare in vendor_name['Name']:
        distance = nltk.edit_distance(compare, vendor)     # Calculate the edit distance between the vendor names
        if distance > 0 and distance < 3:
            print(vendor_name[vendor_name.Name == compare]['Vendor Name'])
            print(vendor_name[vendor_name.Name == vendor]['Vendor Name'])
            print('---------------------------------------')
```

```python
print('It is clear that most of the differences are caused by punctuation mark, \nwe deal with them first.')
```

```python
# Remove the ',' and '.' from the Vendor Name column
df['Vendor Name'] = df['Vendor Name'].astype(str)
df['Vendor Name'] = df['Vendor Name'].replace('[,.]','', regex=True)
```

```python
df['Vendor Name'] = df['Vendor Name'].str.upper().str.strip()       # Convert vendor names to upper cases and strip white spaces
vendor_name = pd.DataFrame(df['Vendor Name'].value_counts(ascending=True))         # Count vendor names
vendor_name['Name'] = vendor_name.index      # Add a new column called Name

q2 = vendor_name['Vendor Name'].quantile(0.5)        # Get the second quartile value in the vendor name count
for vendor in vendor_name[vendor_name['Vendor Name'] < q2]['Name']:
    for compare in vendor_name['Name']:
        distance = nltk.edit_distance(compare, vendor)     # Calculate the edit distance between the vendor names
        if distance > 0 and distance < 3:
            print(vendor_name[vendor_name.Name == compare]['Vendor Name'])
            print(vendor_name[vendor_name.Name == vendor]['Vendor Name'])
            print('---------------------------------------')
```

```python
# Based on the result, we define the vendor name map
vendor_name_map = {}
for vendor in vendor_name['Name']:
    vendor_name_map[vendor] = vendor

vendor_name_map['ST GEORGE SPIRIT INC'] = 'ST GEORGE SPIRITS INC'
vendor_name_map['IMPEX BEVERAGE INC'] = 'IMPEX BEVERAGES INC'

# Map the Vendor Name to its correct values
df['Vendor Name'] = df['Vendor Name'].map(vendor_name_map)
```
