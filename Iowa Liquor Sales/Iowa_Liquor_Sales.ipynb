# Data Cleaning

## Change data type of the Date column

```python
df['Date'] = pd.to_datetime(df['Date'])      # Change data type into datetime
```

## Dealing with missing values

```python

# Define a custom function to calculate mode with missing values
def mode_with_missing(series):
    try:
        return series.mode().iloc[0]
    except IndexError:
        return None

# Group by the Store Number column and aggregate values
aggregate_data = df.groupby('Store Number').agg({
    'Address': mode_with_missing,
    'City': mode_with_missing,
    'Zip Code': mode_with_missing,
    'Store Location': mode_with_missing,
    'County Number': 'first', # Use the first value
    'County': mode_with_missing,
    'Category': mode_with_missing,
    'Category Name': mode_with_missing,
}).reset_index()

# Create a dictionary from the aggregated data
store_mapping = {}
for index, row in aggregate_data.iterrows():
    store_number = row['Store Number']
    store_mapping[store_number] = {
        'Address': row['Address'],
        'City': row['City'],
        'Zip Code': row['Zip Code'],
        'Store Location': row['Store Location'],
        'County Number': row['County Number'],
        'County': row['County'],
        'Category': row['Category'],
        'Category Name': row['Category Name']
    }
store_mapping
```

```python
# Fill missing values based on the dictionary

df['Address'].fillna(df['Store Number'].map(lambda x: store_mapping[x]['Address']), inplace=True)
df['City'].fillna(df['Store Number'].map(lambda x: store_mapping[x]['City']), inplace=True)
df['Zip Code'].fillna(df['Store Number'].map(lambda x: store_mapping[x]['Zip Code']), inplace=True)
df['Store Location'].fillna(df['Store Number'].map(lambda x: store_mapping[x]['Store Location']), inplace=True)
df['County Number'].fillna(df['Store Number'].map(lambda x: store_mapping[x]['County Number']), inplace=True)
df['County'].fillna(df['Store Number'].map(lambda x: store_mapping[x]['County']), inplace=True)
df['Category'].fillna(df['Store Number'].map(lambda x: store_mapping[x]['Category']), inplace=True)
df['Category Name'].fillna(df['Store Number'].map(lambda x: store_mapping[x]['Category Name']), inplace=True)
```

```python
# Check the result of dealing with missing values

num_of_missing_values = df.isnull().sum()         # Count missing values in each feature
percentage_missing = ((num_of_missing_values / df.shape[0]) * 100).apply("{:.2f}%".format)        # Calculate the percentage of missing values

data = {
    'Data Type' : df.dtypes,          # Print data type of each feature
    'Number of Unique Values': df.nunique(),         # Count unique values in each feature
    'Number of Missing Values': num_of_missing_values,
    'Percentage of Missing Values': percentage_missing
}

pd.DataFrame(data)
```

## Check for inconsistency in the City column

```python
df['City'] = df['City'].str.upper().str.strip()       # Convert city names to upper cases and strip white spaces
city = pd.DataFrame(df['City'].value_counts(ascending=True))         # Count cities
city['Name'] = city.index      # Add a new column called Name
city
```

```python
q1 = city['City'].quantile(0.25)        # Get the second quartile value in the city count
for city_name in city[city['City'] < q1]['Name']:
    for compare in city['Name']:
        distance = nltk.edit_distance(compare, city_name)     # Calculate the edit distance between the cities
        if distance > 0 and distance < 3:
            print(city[city.Name == compare]['City'])
            print(city[city.Name == city_name]['City'])
            print('---------------------------------------')
```

```python
# Based on the result, we define the city map
city_map = {}
for city_name in city['Name']:
    city_map[city_name] = city_name

city_map['ALVORD'] = 'WALFORD'
city_map['CLEARLAKE'] = 'CLEAR LAKE'
city_map['GRAND MOUNDS'] = 'GRAND MOUND'
city_map['OTUMWA'] = 'OTTUMWA'

df['City'] = df['City'].map(city_map)
```

## Check whether records with same Store Number have the same information

```python
consistency_check_columns = ['Store Name', 'Address', 'City', 'Zip Code', 'Store Location', 'County']
for column in consistency_check_columns:
    df[column] = df[column].str.upper()

# Group by 'Store Number' and check the number of unique values in each column
grouped = df.groupby('Store Number').agg({
    'Store Name': 'nunique',
    'Address': 'nunique',
    'City': 'nunique',
    'Zip Code': 'nunique',
    'Store Location': 'nunique',
    'County Number': 'nunique',
    'County': 'nunique',
    'Latitude': 'nunique',
    'Longitude': 'nunique'
})

grouped
```

```python
# Check if all columns have a count of 1 within each group
consistent = grouped.eq(1).all(axis=1)
```

```python
# Records with the same Store Number and Address should have the same Latitude and Longitude

# Group by 'Store Number' and 'Address' to identify inconsistencies
inconsistent_groups = df.groupby(['Store Number', 'Address'])[['Latitude', 'Longitude']].nunique().reset_index()

# Filter groups with more than one unique value for 'Latitude' or 'Longitude'
inconsistent_groups = inconsistent_groups[(inconsistent_groups['Latitude'] > 1) | (inconsistent_groups['Longitude'] > 1)]

# Update the DataFrame to make 'Latitude' and 'Longitude' consistent within each group
for index, row in inconsistent_groups.iterrows():
    store_number = row['Store Number']
    address = row['Address']
    consistent_latitude = df[(df['Store Number'] == store_number) & (df['Address'] == address)]['Latitude'].mode().iloc[0]
    consistent_longitude = df[(df['Store Number'] == store_number) & (df['Address'] == address)]['Longitude'].mode().iloc[0]
    df.loc[(df['Store Number'] == store_number) & (df['Address'] == address), 'Latitude'] = consistent_latitude
    df.loc[(df['Store Number'] == store_number) & (df['Address'] == address), 'Longitude'] = consistent_longitude
```

```python
# Double check the result of dealing with inconsistency
consistency_check_columns = ['Store Name', 'Address', 'City', 'Zip Code', 'Store Location', 'County']
for column in consistency_check_columns:
    df[column] = df[column].str.upper()

# Group by 'Store Number' and check the number of unique values in each column
grouped = df.groupby('Store Number').agg({
    'Store Name': 'nunique',
    'Address': 'nunique',
    'City': 'nunique',
    'Zip Code': 'nunique',
    'Store Location': 'nunique',
    'County Number': 'nunique',
    'County': 'nunique',
    'Latitude': 'nunique',
    'Longitude': 'nunique'
})

grouped
```
